```python
import pandas as pd
# 데이터 불러오기
dataset = pd.read_csv("pulsar_stars.csv")
# 데이터 확인
dataset.head()
dataset.info()
```

```python
# check distribution of target_class column

dataset['target_class'].value_counts()
X = dataset.drop(['target_class'], axis=1)

y = dataset['target_class']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 32190403)
```

```python
print("특성1 범위: ", "[", min(X_train[:, 0]), ",", max(X_train[:, 0]), "]")
print("특성2 범위: ", "[", min(X_train[:, 1]), ",", max(X_train[:, 1]), "]")
print("특성3 범위: ", "[", min(X_train[:, 2]), ",", max(X_train[:, 2]), "]")
print("특성4 범위: ", "[", min(X_train[:, 3]), ",", max(X_train[:, 3]), "]")
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
print("표준화된 특성1 범위: ", "[", min(X_train_std[:, 0]), ",", max(X_train_std[:, 0]), "]")
print("표준화된 특성2 범위: ", "[", min(X_train_std[:, 1]), ",", max(X_train_std[:, 1]), "]")
print("표준화된 특성3 범위: ", "[", min(X_train_std[:, 2]), ",", max(X_train_std[:, 2]), "]")
print("표준화된 특성4 범위: ", "[", min(X_train_std[:, 3]), ",", max(X_train_std[:, 3]), "]")
```

```python
from sklearn.svm import SVC

svm_model = SVC(kernel='rbf', C=8, gamma=0.1)

svm_model.fit(X_train_std, y_train)
```

```python
y_pred = svm_model.predict(X_test_std)
```

```python
import numpy as np
```

```python
print("prediction accuracy: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
svm_model = SVC(kernel='rbf', C=10, gamma=0.1)

svm_model.fit(X_train_std, y_train)
y_pred = svm_model.predict(X_test_std)
print("prediction accuracy: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
svm_model = SVC(kernel='linear', C=100, gamma=0.1)

svm_model.fit(X_train_std, y_train)
y_pred = svm_model.predict(X_test_std)
print("prediction accuracy: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
from sklearn.metrics import roc_curve

# compute ROC AUC

from sklearn.metrics import roc_auc_score

ROC_AUC = roc_auc_score(y_test, y_pred)

print('ROC AUC : {:.4f}'.format(ROC_AUC))
```

```python
# f1-score
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)
print("f1 score: ",f1)
```

```python
# AUPRC score
from sklearn.metrics import average_precision_score, precision_recall_curve
from sklearn.metrics import auc, plot_precision_recall_curve
# Data to plot precision - recall curve
precision, recall, thresholds = precision_recall_curve(y_test, y_pred)

# Use AUC function to calculate the area under the curve of precision recall curve
auc_precision_recall = auc(recall, precision)
print("AUPRC score: ",auc_precision_recall)
```

```python
# MCC score
from sklearn.metrics import matthews_corrcoef
matthews_corrcoef(y_test, y_pred)
```
